```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from sklearn import preprocessing
```

```python
df = pd.read_csv("iris.data", header = None)
df
```

```python
x = df.iloc[:,:4].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(x)
X
```

```python
x1 =X[:,0]
y1 = X[:,1]
z= X[:,2]
p= X[:,3]
x1
```

```python
label_encoder = preprocessing.LabelEncoder()
df[4]= label_encoder.fit_transform(df[4])
df
```

```python
x_mean = 0;
y_mean = 0;
z_mean=0;
p_mean=0;
n = len(x1)
for i in range(n):
    x_mean += x1[i]
    y_mean += y1[i]
    z_mean += z[i]
    p_mean += p[i]
x_mean = x_mean / n
y_mean = y_mean / n
z_mean = z_mean / n
p_mean = p_mean / n
```

```python
cov_xy = 0;
cov_xx = 0;
cov_yy = 0;
cov_zz=0;
cov_pp=0;
cov_xz=0;
cov_xp=0;
cov_yz=0;
cov_yp=0;
cov_zp=0;


for i in range(n):
    cov_xy += ( (x1[i] - x_mean) * (y1[i] - y_mean) )
    cov_xx += ( (x1[i] - x_mean) * (x1[i] - x_mean) )
    cov_yy += ( (y1[i] - y_mean) * (y1[i] - y_mean) )
    cov_zz += ( (z[i] - z_mean) * (z[i] - z_mean) )
    cov_pp += ( (p[i] - p_mean) * (p[i] - p_mean) )
    cov_xz += ( (x1[i] - x_mean) * (z[i] - z_mean) )
    cov_xp += ( (x1[i] - x_mean) * (p[i] - p_mean) )
    cov_yz += ( (y1[i] - y_mean) * (z[i] - z_mean) )
    cov_yp += ( (y1[i] - y_mean) * (p[i] - p_mean) )
    cov_zp += ( (z[i] - z_mean) * (p[i] - p_mean) )
cov_xy = cov_xy / ( n - 1 )
cov_xx = cov_xx / ( n - 1 )
cov_yy = cov_yy / ( n - 1 )
cov_zz = cov_zz / ( n - 1 )
cov_pp = cov_pp / ( n - 1 )
cov_xz = cov_xz / ( n - 1 )
cov_xp = cov_xp / ( n - 1 )
cov_yz = cov_yz / ( n - 1 )
cov_yp = cov_yp / ( n - 1 )
cov_zp = cov_zp / ( n - 1 )
```

```python
C = [ [cov_xx, cov_xy,cov_xz,cov_xp], [cov_xy, cov_yy,cov_yz,cov_yp],[cov_xz, cov_yz,cov_zz,cov_zp],[cov_xp, cov_yp,cov_zp,cov_pp], ]
```

```python
values, vectors = np.linalg.eig(C)
values[:5]
```

```python
projected_1 = X.dot(vectors.T[0])
projected_2 = X.dot(vectors.T[1])
res = pd.DataFrame(projected_1, columns=['PC1'])
res['PC2'] = projected_2
res['Y'] = y
res.head()
```
